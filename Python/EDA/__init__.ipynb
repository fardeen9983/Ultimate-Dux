3. Features are also known as variable, input/ dependent variable

```python
# Import required libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# iris dataset
# The last column is the label for the flower with given features
# We need to classify a flower into one of the three labels given these four features
iris = pd.read_csv("iris.data")
iris
```

```python
# Columns in the dataset
iris.columns
```

```python
# Find the no of datapoints given for each class
# Here each class has equal number of datapoint
# Such a dataset is called balanced while rest are known as imbalanced datasets
iris.species.value_counts()
```

### 2-D scatter plot 

```python
# Plot a 2-D scatter plot for the sepal features
# Define the axises and the type of plot 
iris.plot(kind='scatter',x='sepal_length',y='sepal_width')
plt.show()
```

### 2-D color coded scatter plot ro differentiate differernt iris classes

```python
# Set plot style
sns.set_style("whitegrid")
# Set plot properties using FacetGrid : Multi-plot grid for plotting conditional relationships
sns.FacetGrid(iris,hue="species",height=4)\
    .map(plt.scatter,"sepal_length","sepal_width")\
    .add_legend()
# Plot the graph
plt.show()
```

**Observation : Sepal length vs width**

1. By using sepal length and width we can easily differentiate setosa flowers from the rest 
2. It is difficult to differentiate between versicolor and virginica flowers from this feature set

**Conclusion :**
Sepal length and width makes setosa flowers linearly separable from the others (by drawing a line between them)  

### Plot pair plots on datasets to find the best features to identify the various classes

```python
sns.set_style("whitegrid")
# Plot the Pair plot
sns.pairplot(iris,hue="species",height=3)
# The output is a a matrix of plot with axises as the features
# The diagonal plots are the pair plots
plt.show()
```

### Observation :
**Best feature pairs to identify different iris species : Petal width vs Petal length** 

Constraints on these two features will help us to classify flowers easily

Pair plots are easy to understand when number of features is small. It becomes increasing difficult to grasp them as dimensionality increases
